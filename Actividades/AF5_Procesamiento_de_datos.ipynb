Esta celda es para implementar el archivo robot_sensors.csv, que descargamos previamente

```python
data = pd.read_csv("robot_sensors.csv")
```

Esta celda es para cargar el dataset

```python
data.head()
```

En esta celda se cargó el dataset utilizando la librería Pandas para analizar la información

```python
data = data.dropna()
```

En esta celda se eliminan los valores nulos

```python
data = data[data["Velocidad (cm/s)"] >= 0]
```

En esta celda se filtran los valores nulos y los valores con características como los negativos

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

data_normalized = pd.DataFrame(
    scaler.fit_transform(data),
    columns=data.columns
)

data_normalized.head()
```

En esta celda se convierten todos los valores entre 0 y 1

```python
plt.figure()

plt.plot(data_normalized["Tiempo (s)"], data_normalized["Distancia (cm)"])

plt.xlabel("Tiempo (s)")
plt.ylabel("Distancia normalizada")

plt.title("Distancia VS tiempo")

plt.show()
```

En esta celda pedimos hacerle una gráfica de distancia contra tiempo

```python
plt.figure()

plt.plot(data_normalized["Tiempo (s)"], data_normalized["Velocidad (cm/s)"])

plt.xlabel("Tiempo (s)")
plt.ylabel("Velocidad normalizada")

plt.title("Velocidad VS tiempo")

plt.show()
```
